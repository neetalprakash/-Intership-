```python
rating=[]
for i in driver.find_elements(By.XPATH,'//div[@class="ipl-rating-widget"]'):
    rating.append(i.text.replace("\nRate"," "))
rating
```

```python
votes=[]
for i in driver.find_elements(By.NAME,'nv'):
    votes.append(i.text)
votes
```
