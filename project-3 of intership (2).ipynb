search results and save it in a data frame and csv. In case if any product has less than 3 pages in search

Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and

```python
driver.get("https://www.amazon.com/s?k=guitars&crid=21D6YW54NQFBD&sprefix=guitars%2Caps%2C584&ref=nb_sb_noss_1")
```

```python
comments=[]
for i in driver.find_elements(By.XPATH,'//yt-formatted-string[@class="style-scope ytd-comment-renderer"]')[0:500]:
    comments.append(i.text)
    
comments
```

```python
len(comments)
```

```python
comment_upvote=[]
for i in driver.find_elements(By.CLASS_NAME,"style-scope ytd-comment-action-buttons-renderer")[0:500]:
    comment_upvote.append(i.text.replace('\n',' '))
    
comment_upvote
```

```python
len(comment_upvote)
```

```python
posted_time=[]
for i in driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')[0:500]:
    posted_time.append(i.text)
    
posted_time
```
