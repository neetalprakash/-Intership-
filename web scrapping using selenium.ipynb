Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.

```python
#connecting driver

driver=webdriver.Chrome(r"chromedriver.exe")
driver
```

```python
#opening nokri.com page for scrapping 

driver.get("https://www.naukri.com/data-analytics-jobs-in-bangalore?k=data%20analytics&l=bangalore")
```

```python
#scrapping the job_title data from the nokri.com page

job_title=[]
title_tags=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
job_title
```

```python
#scrapping the location data from the nokri.com page

job_location=[]
location_tags=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)
job_location
```

```python
#scrapping the compny name from the nokri.com page

company_name=[]
company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')

for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)
company_name
```

```python
#scrapping the  experience_required from nokri.com page

experience_required=[]
experience_tags=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')

for i in experience_tags[0:10]:
    experience=i.text
    experience_required.append(experience)
    
experience_required
    

```

```python
#making dataframe of data analytics dataset

DATA_OF_DATA_ANALYTICS=pd.DataFrame({'JOB_TITLE':job_title,'JOB_LOCATION':job_location,'COMPANY_NAME':company_name,'EXPERIENCE_REQUIRED':experience_required})
DATA_OF_DATA_ANALYTICS
```
